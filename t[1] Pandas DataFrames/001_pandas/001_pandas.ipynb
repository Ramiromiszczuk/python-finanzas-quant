### Importacion de un XLS con Pandas


```python
# En caso de importar un XLS como dataframes es un poco mas complejo
# El objeto XLS es mas completo que un archivo CSV ya que por ejemplo tenemos
# multiples hojas de calculo en el mismo libro o archivo, por lo tanto,
# debemos pasarselo como argumento para que pandas sepa que hoja queremos 
# tomar los datos. Si no le pasamos el argumento sheet_name, nos va a traer
# lo que encuentre en la primera hoja
data = pd.read_excel("AAPL.xlsx", sheet_name="Hoja1")
data
```

```python
# Traigamos la misma tabla, pero indicando que el indice sea la fecha
# en lugar de un indice numerico, para ello usamos el atributo index_col
data = pd.read_excel('AAPL.xlsx', sheet_name='Hoja1', index_col="timestamp")
data
```

```python
# Tambien podemos traer solamente una determinada cantidad de columnas
# que necesitemos, para ello usamos el atributo usecols y le pasamos
# la lista de columnas que queremos traer como argumento.
columnas = ["timestamp", "open", "close"]
data = pd.read_excel('AAPL.xlsx', sheet_name='Hoja1', index_col="timestamp", usecols=columnas)
data
```

### Configurando la salida

Cuando mandamos a imprimir o mandamos a pantalla una variable que es un dataFrame con muchas filas, por defecto Pandas nos muestra 10 filas en total, las primeras 5 y las ultimas 5 filas, algo similar hace con las columnas, esto es un valor default que se puede configurar.
Vamos a limitar la cantidad de filas o de colunas que quiero que me muestre en pantalla. Para ello usamos el objeto pd que es un objeto de pandas vacio antes de llenarlo con la tabla del Excel.


```python
pd.options.display.max_rows = 4
pd.options.display.max_columns = 6

data = pd.read_excel('AAPL.xlsx', sheet_name='Hoja1')
data
```

Si queremos que el numero de filas o columnas a mostrar sea limitado, es decir que muestre siempre todo, lo configuramos como _None_


```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 6
data
```

### Opciones de Formato

Podemos modificar el formato de salida, mas que nada la cantidad de decimales que muestra, a veces puede ser medio denso que muestre muchos decimales y facilita la vista que muestre 2 o 1 o ninguno.

Cabe destacar que redondea los valores, es decir que el numero 3.45678 si lo mostramos con 3 decimales lo mostrara como 3.457

Si le indicamos un numero de decimales mayor a la que tienen todas las columnas, no hara nada.


```python
pd.options.display.precision = 4
data
```

Mas opciones de formato


```python
pd.options.display.precision = 2
data
```
